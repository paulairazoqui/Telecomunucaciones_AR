## 2. Distribución Temporal
Se analiza cómo evolucionan los accesos a lo largo del tiempo (Año y Trimestre) agrupando por Año.

```python
internet_accesos.groupby('Año')['total'].sum().plot(kind='bar', title='Total de Accesos por Año')
```

Dado que el año 2024 tiene solo dos trimestres, se ajusta el análisis para considerar esa diferencia y así obtener una comparación más justa con los años anteriores.

**Ajuste del Gráfico**
Se calcula un promedio trimestral de los accesos para cada año y lo graficaremos. Esto hará que los datos del 2024 sean proporcionales a los años completos.

```python
# Crear una nueva columna con los accesos promedio por trimestre
internet_accesos['Accesos_promedio_trimestre'] = internet_accesos['total'] / internet_accesos['Trimestre']

# Agrupar por año y calcular el promedio anual (ajustado por trimestres)
accesos_promedio_por_año = internet_accesos.groupby('Año')['Accesos_promedio_trimestre'].mean()

# Graficar
accesos_promedio_por_año.plot(kind='bar', figsize=(10, 6), title='Accesos Promedio por Año (ajustado por trimestre)')
plt.ylabel('Accesos promedio')
plt.show()
```

### Conclusiones

1. **Tendencia Creciente**:
   - Desde 2014 hasta 2023, se observa una tendencia general al alza en el número total de accesos a Internet, lo que refleja una mayor adopción tecnológica y crecimiento en la conectividad.

2. **Pico en 2023**:
   - El año 2023 muestra el valor más alto en la serie, indicando que fue un año clave en términos de expansión de accesos.

3. **Disminución en 2024**:
   - El 2024 muestra un descenso abrupto en el gráfico, pero esto se debe a que los datos solo abarcan los dos primeros trimestres del año. Este hecho debe tenerse en cuenta antes de sacar conclusiones sobre una posible disminución en la conectividad.

4. **Estabilización de Crecimiento**:
   - A partir de 2020, el crecimiento parece estabilizarse, con valores similares entre 2020 y 2022, posiblemente influenciado por la saturación del mercado o un aumento previo más acelerado debido a factores como la pandemia y el trabajo remoto.

5. **Impacto de Políticas y Tecnología**:
   - La aceleración del crecimiento en años recientes podría estar influenciada por políticas de conectividad, avances tecnológicos como el aumento de fibra óptica, y un cambio en las necesidades digitales de la población.

6. **Importancia del Ajuste por Trimestres**:

Este gráfico proporciona un panorama útil para entender la evolución de la conectividad a lo largo de los años, destacando tanto las tendencias como la importancia de ajustar los datos para análisis precisos.

## 6. Análisis de Categorías
 **Contexto**:
**ADSL** (Asymmetric Digital Subscriber Line) es una tecnología de transmisión de datos que permite acceder a Internet utilizando las líneas telefónicas de cobre existentes. Es una forma de conexión de banda ancha que fue ampliamente utilizada antes de la aparición de tecnologías más avanzadas como Cablemodem y Fibra Óptica.

---

### Características principales de ADSL:
1. **Asimetría**:
   - La velocidad de descarga (descendente) es significativamente mayor que la velocidad de subida (ascendente). Esto se diseñó para satisfacer las necesidades típicas de los usuarios, que generalmente descargan más datos (navegar, ver videos, etc.) que los que suben.

2. **Uso de Líneas Telefónicas**:
   - ADSL utiliza la infraestructura de las líneas telefónicas de cobre ya instaladas, lo que facilitó su implementación masiva en zonas urbanas y rurales.

3. **Conexión Simultánea**:
   - Permite utilizar el teléfono y acceder a Internet al mismo tiempo, gracias a la separación de frecuencias.

4. **Velocidades**:
   - Velocidades típicas:
     - **Descarga**: Entre 1 Mbps y 24 Mbps (dependiendo de la calidad de la línea y la distancia a la central telefónica).
     - **Subida**: Generalmente entre 128 Kbps y 1 Mbps.
   - La velocidad depende de la distancia entre el usuario y la central telefónica: cuanto más lejos esté, menor será la velocidad.

5. **Costos y Disponibilidad**:
   - Fue una tecnología económica y fácil de implementar, lo que la convirtió en la opción más popular en muchas regiones durante años.

---

### Ventajas:
- Amplia disponibilidad, ya que utiliza la infraestructura telefónica existente.
- Costos relativamente bajos en comparación con otras tecnologías más avanzadas en su momento.
- Permite conexión simultánea de voz e Internet.

---

### Desventajas:
- Limitaciones en la velocidad de subida y descarga en comparación con tecnologías más modernas como Fibra Óptica.
- Pérdida de rendimiento a medida que aumenta la distancia entre el usuario y la central telefónica.
- No es adecuada para aplicaciones que requieren altas velocidades de subida (como transmisiones en vivo o cargas de datos masivas).

---

### Estado Actual:
- Aunque sigue siendo utilizada en algunas áreas rurales y con infraestructura limitada, ADSL está siendo reemplazada por tecnologías más avanzadas como Cablemodem, Fibra Óptica y redes inalámbricas de alta velocidad.

---

**Conclusión**:
ADSL fue una tecnología revolucionaria en su momento, pero actualmente está quedando obsoleta frente a opciones que ofrecen mayores velocidades y mejor calidad de servicio.
